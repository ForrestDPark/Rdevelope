#  가설검정
### 카이제곱 검정
: 두 범주형 변수가 서로 상과나이 있는지 판단하는 통계적 검정 방법.
예) 학력, 성별, 직업의 만족도


### child 별 장난감 보유 현황
두개의 데이터가지고 통계를 한다. 3개이상은 잘쓰지 않는다. 
귀무가설이란 일반적인 상식같은것, 분석할 가치도 없는데? 하는게 귀무가설. 사람은 숨을 쉬어야 살수가 있는가? 아무 의미가 없지만 참인 가설.
당연한 이야기 , 계는 귀무가설이야 = 할게 없는 놈이야. 
child1 돌아갈 귀에 없을무 무로 돌아간다....


- 귀무가설 : child1 과 child2의 장난감 보유수에 대한 만족도는 차이가 없다. 
- 대립가설 : child1 과 child2의 장난감 보유수에 대한 만족도의 차이가 있다. ( 정답이 아님. 데이터에 따라 달라짐.)


```r
# child 별 장난감 보유 현황
child1 <- c(5,11,1)
child2 <- c(4,7,3)

# 분석을 위해 데티어 합치기 
Toy <- cbind(child1, child2); Toy
```

```r
# 행의 이름  삽입
rownames(Toy)<- c('car','truck','doll'); Toy
```

```r
# 카이제곱 검정
chisq.test(Toy) # p-value :0.4219
```

### 결론 
p-value = 0.4219 > 0.05 ( 유의수준 : 95% 의 신뢰구간 ) --> 귀무가설임. 
child1 과 child2의 데이터는 통계적으로 큰 차이가 없다. 

---
p value 는 0에 가까울수록 좋다. ( 100% 신뢰할수있는 데이터이다.)
미리 정해진 유의수준(일반적으로 0.05) 보다 작으면 대립가설을 채택하고 크면 귀무가설을 채택한다. 

df =2 : 검정에서 사용된 자유도는 2이다. 
카이제곱 검정의 자유도는 3개의 범주가 있다면 자유도는 3-1=2 가 됩니다.  



### Warning

통계 데이수 최소 : 5개, 머신러닝은 150개 
데이터량이 적은경우 warning 발생
--> 피셔검정 : 표본수가 적거나 데이터의 분포가 치우친 경우



```r
# 피셔검정
fisher.test(Toy)
```

p -value  =0.5165 > 0.05 귀무가설 
대립가설 : 대립가설은 'two.sided'로 설정되어있어 양측검정을 나타낸다. 분포에서 양쪽을 잘라내고 쓰겠다. 


```r
# 상관계수 를 보면된다.
cor(child1,child2)
```

t-검정



```r
# 건전지의 수명은 1000시간이다. <- 귀무가설
# 대립가설 : 건전지의 수명은 1000시간이 아니다. 1000이 안될수도 있고 넘을수도 있다. 

#광고회사한테 베터리 데이터를 받았다. 
# 데이터의 분포가 정규분포인가 확인 (Shapiro - Wilk TEST)
# 귀무가설 : 자료가 정규분포를 따른다. 
# 대립가설 : 자료가 정규분포를 따르지 않는다. 
bat <-c(980,1008,968,1032,1012,1002,996,1017) # 베터리 수명시간. 
shapiro.test(bat)
```

 p-value = 0.9469 >0.05 -> 귀무가설 => 정규분포이다.  <br>
 W = W = 0.97706 값이 1에 가까울수록 정규성에 가깝다. 

```r
# t- 검정
t.test(bat, mu =1000, alternative ="two.sided") 
# mu : 비교하는 대상의 평균
# alternative = 
    # - two.sided : 데이터가 평균과 다르다. (작거나 크다) 
    # - greater : 데이터가 평균보다 크다. 
    # - less : 데이터가 평균보다 작다. 
```

 p-value = 0.8032 >0.05 -> 귀무가설 : 광고는 허위 광고가 아니었다. 

```r
# 어떤 학급의 수학 점수 평균 : 55점
# 0 교시 수업을 한 후에 학생들의 성적이 올랐을까?
# 귀무가설 : 성적이 오르지 않았다. 
# 대립가설 : 성적이 올랐다. 
```

```r
exam<-read.csv("../Data/exam.csv")
head(exam)
# 정규 분포인지 확인 샤피로
# 귀무가설 : 정규분포이다. 
# 대립가설 : 정규분포가아니다. 


shapiro.test(exam$score)
# p =0.1058 >0.05 귀무가설이다 -> 정규 분포이다. 

# 귀무가설 : 성적이 오르지 않았다. -> 비교치( Mu =55)
# 대립가설 : 성적이 올랐다. ( alternative = greater)
t.test(exam$score, mu= 55, alternative = "greater")

mean(exam$scor)
par(bg='white')
boxplot(exam$score)
stem(exam$score)
```

# 상관관계 

```r
y <- c(1,2,3,4)
z <- c(0,5,7,9)
z1 <- c(5,0,7,9)
cor(y,z)
cor(y,z1)
# 순서가 달라져도 상관관계가 달라진다. 
#  x, y 를 정렬해서 하면 무조건 상관관계가 높아질수밖에 없다. 이것은 데이터 조작이다. !
```

```r
# 피어슨 상관계수
cor(y,z, method ='pearson') # default 가 pearson 임 5개 이상 데이터에만 적용가능 
# 스피어맨 상관계수( 데이터 적을때 )
cor(y,z, method = 'spearman') # 1 이 나옴. 
# 역상관도 상관이다. 반비례 관계 임. 
```

# 콜모고로프-스미노프 검정 KS TEST
- 주어진 2개의 데이터가 같은 분포를 갖는지를 검정



```r
# 귀무가설 : 두개의 데이터의 분포가 같다. 
# 대립가설 : 두개의 데이터의 분포가 다르다. 
```

```r
x<- rnorm(50);
y<- runif(50)
ks.test(x,y) # 4.048e-08 > 0.05 대립가설임. : 분포가 다르다 .
par(bg ='white')
plot(x)
plot(y)# 데이터 범위가 다르다. 분포가 비슷해도 아무소용없다. 

```

# 부호검정
: 2개의 데이터 사이에 차이가 잇는지 검정하는 것


```r
# 식사전과 식사후의 음료수 맛에 대한 평가. 
x <- c(4,1,1,4,3,3,2,5,3,3) # 식사 전 맛에 대한 평가
y  <-c(1,1,3,2,5,1,4,4,3,1) # 식사 후 맛에 대한 평가
```

```r
# 귀무가설 : 유의미한 차이가 없다. 
# 대립가설 : 유의미한 차이가 있다. 
binom.test(c(length(x[x>y]),length(x[x<y]))) # p:0.7266 >0.05 귀무가설 
```

# 비율 검정. 
: 2개의 데이터 사이에 비율의 차이가 잇는지를 검정. 


```r
# 맥주를 좋아하시나요 ? 라는 질문에 서울에서는 400명 중 360명 , 부산에서는 200명중 136명이 좋아한다고 답했다면
# 둘의 비율 차이가 잇따고 할 수 가 있는가? (adp 문제임.)
# 귀무가설 : 유의미한 차이가 없다. 
# 대립가설 : 유의미한 차이가 있다. 

hite <- c(360,136)
sample <- c(400,200)
prop.test(hite, sample) # 4.207e-11 <0.05 : 대립가설 -> 서울이 부산보다 더 맥주를 좋아한다.../0.9와 0.68 차이 비교함.
```

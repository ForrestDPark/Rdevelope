```r
# 난수 (random number) : 0 < x < 1 0과 1 은 판단지수가 되기때문에 뺀다. 0과 1 사이에 어떤수가 난수이다. 
# 분석에서 어떤 가설을 증명하고 싶을때 그 데이터가 없을때 난수를 사용하여 데이터를 만든다. 
# 난수 10개를 발생
runif(19)
```

```r
# 소수점 이하 버림
trunc(100.5)
```

```r
# 100 보다 적은 난수 정수 10개 만들기
trunc(100*runif(10))
```

```r
# 로또 645 숫자 생성하기 1~45
# help(runif)
# 머신러닝 최소 데이터 150 딥러닝 최소 개수 10만개  이거 모으거나 난수로 데이터 수집하거나.. 
trunc((runif(6, min= 1, max=45)))
```

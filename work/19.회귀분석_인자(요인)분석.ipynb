# 다중 회귀 분석 

```r
par(bg = 'white')
plot(attitude)
```

```r
# rating 을 제외한 변수들이 rating 에 어떤 영향을 미치는지 회귀 분석
model <-lm(rating~ ., data = attitude)
model 
```

 Rating = 0.61319 *complaints + -0.07305 + privileges +0.32033*learning +0.08173*raises

```r
summary(model)
```


Adjusted R-squared:  0.6628 

```r
# 상관계수 
cor(attitude)
```

```r
# 공식이 너무길어서 좀 줄이고 싶다. 컴팩한 공식을 만들고 싶다. 파이썬에는 회귀 분석에 대해서 이렇게 디테일하게 없다. 
# 예측력을 높이기 위해 상대적으로 기여도가 낮은 변수를 제거 
# 기여도. 를 반영해서 계수를 제거해주는 함수  
# 예측할 때 필요한 컬럼과 분석할 때 필요한 컬럼이 다르다.
reduced <- step (model , direction = "backward") # 낮은 애들을 다 제거하면서 거꾸로 와라.
```

```r
summary(reduced)
```

Before -> Adjusted R-squared:  0.6628 
After -> Adjusted R-squared:  0.6864 

```r
reduced
```

# 교통사고 예측하기 ( 회귀식과 RandomForest 예측비교 )

```r
# 교통사고 데이터 불러오기
Sys.setlocale(category = "LC_ALL", locale = "ko_KR.UTF-8") # 한글 설정

seoul <- read.csv("../Data/newSeoul2019.csv")
head(seoul)
```

```r
# 종로구 데이터셋 만드릭 
JongRo <- subset(seoul, 자치구명 =="종로구")
nrow(JongRo)

```

```r
## row names 정리하기 
# rownames(JongRo)
rownames(JongRo) <- 1:nrow(JongRo)
```

```r
head(JongRo)
str(JongRo)
```

```r
# 종로구의 발생건수와 부상자수의 상관계수
cor(JongRo$`발생건수`,JongRo$`부상자수`) # 92% 이니까 회귀선을 그릴수 있겠다. 
cor(JongRo[4:6])
```

```r
options(warn =-1)
par(bg ='white', family = "AppleGothic")
plot(JongRo$`발생건수`,JongRo$`부상자수`,
    main =" 종로구의 발생건수와 부상자수의 산포도",
    ylab ="부상자수 ",
    xlab ="발생건수"
)
r <-lm(JongRo$`부상자수`~JonRo$`발생건수`)

abline(r,col="red")
```

```r
summary(r)
```

```r
# 회귀식으로 예측 : 발생건수 100건 일 경우 예상 부상자 수는?
# y : 부상자 , x : 발생건수 
x <- 100
y <- 1.35993*x +2.88612
y
```

```r
library(randomForest)
```

```r
# 랜덤포레스트 예측 : 발생건수가 100일 경우의 부상자수 ?
rf <- randomForest(부상자수 ~ 발생건수, data =JongRo, ntree =100)
predict(rf,list(발생건수 =100) ) # 종로는 데이터 프레임이었다. rf한테 물어볼때도 똑같은 타입으로 물어봐야한다. 
```


### 발생건수가 145건 일경우 회귀식과 RandomForest를 통한 부상자수를 예측하시오 

```r
n_acc <-c(145,150,160)

compare_acc<- function(n_acc){
    rf_pred <-predict(rf,list(발생건수=n_acc))
    regresesion_pred <- 1.35993*(x<-n_acc) +2.88612;
    cat(paste("random forest predict : ",rf_pred,"\n",sep=""))
    cat(paste("regression : ",regresesion_pred,"\n", sep=""))
}
compare_acc(n_acc)
# 머신러닝은 데이터에 기반하기때문에 데이터 가 근처에 없는 관측값에 대한 예측은 안된다. 
# 그리고 데이터의 양이 150개가 안된다. 
```

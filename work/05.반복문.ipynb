```r
gugu2 <- function(dan,mul,operator){

    mul_array<-1:mul
    for(i in mul_array){
        cat( dan,operator,i,"=",2*i,'\n')
    }
}
gugu2(3,"x")
```

```r
gugu2 <- function(dan,mul,operator){

    mul_array<-1:mul
    
    for(i in mul_array){
        switch(operator, 
                "x" = cat( dan,"X",i,"=",dan*i,'\n'),
                "+" = cat( dan,"+",i,"=",dan+i,'\n'),
                "-" = cat( dan,"-",i,"=",dan-i,'\n'),
                "%" = cat( dan,"-",i,"=",dan%%i,'\n'),
                "/" = cat( dan,"/",i,"=",dan/i,'\n') )
        
    }
}
gugu2(3,5,"/")


## 구구단 전체 출력하기 

## 구구단 중 짝수단과 짝수 곱해지는 수만 출력하기 

evenDan<- seq(2,9,2); evenMul <-seq(2,9,2)
evenMul
```

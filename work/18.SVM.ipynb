```r
#학습용 , 검증용 분류 
samp <- createDataPartition(iris$Species, p =0.7, list =F)
#학습용
data.tr<- iris[samp,]
# 검증용
data.test<- iris[-samp,]
```

```r
# 문제와 정답 만들기 
x <- subset(data.test, select =-Species) 
y <- data.test$Species
```

```r
# 학습 시키기
model <- svm(Species~ .,data =data.tr)
```

```r
# 검증하기 
pred <- predict(model, x)
table(pred,y); mean(pred==y)

```

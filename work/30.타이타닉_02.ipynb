```r
train_df <- read.csv("../Data/Titanic_train.csv")
head(train_df)
```

```r
train_df_T<-subset( train_df, !is.na(train_df$Age))

# 성별을 숫자로 전환
train_df_T$SexCode[train_df_T$Sex =="male"]<-0
train_df_T$SexCode[train_df_T$Sex =="female"]<-1

# Embarked one hot encoding
train_df_T$Embarked_C <- as.integer(train_df_T$Embarked =="C")
train_df_T$Embarked_S <- as.integer(train_df_T$Embarked =="S")
train_df_T$Embarked_Q <- as.integer(train_df_T$Embarked =="Q")
```

```r
options(repr.plot.width =16, repr.plot.height =16)
options(scipen =100)
target_col <-c("Survived","SexCode","Pclass","Fare")

cat("data rows (Na 제외) :",nrow(train_df_T))
train_df_T<- train_df_T[,target_col]
cat("-------")

special_case <- rownames(subset(train_df_T, SexCode==0&Pclass==3 & Survived ==1))



train_df_T$Survived<- as.factor(train_df_T$Survived)
#nrow(train_df)
cat("-------")


pFactor <-seq(1,2,0.1) # 1.7 일때 5.8
sFactor <-seq(0,1,0.1) # sFactor 1 일때 
findSPFactor <- function(pFactor,sFactor,target_cor){
    for( p in pFactor){

        for(s in sFactor){
        X<-c()
        c<-c()
        for (i in 1:nrow(train_df_T)){
            c<-(ifelse(train_df_T$SexCode[i]==1,s,0.1)) * ifelse(train_df_T$Pclass[i]==1,p,1)
            #cat(c, "\n")
            X<- append(X,c)
        }
        train_df_T$X<-X

        # if(rfFunc(train_df_T,1,130,5)>0.75){
                
        #     cat("\ns : ",s,"p : ",p,cor(as.numeric(train_df_T$Survived),train_df_T$X),"\n")

        # }

        #cat("pFactor : ",p,cor(as.numeric(train_df_T$Survived),train_df_T$X),"\n")
        if (cor(as.numeric(train_df_T$Survived),train_df_T$X)>target_cor){
            cat("s : ",s,"p : ",p,cor(as.numeric(train_df_T$Survived),train_df_T$X),"\n")

        }

        
    }

    }
    
}
#findSPFactor(pFactor,sFactor,0.57)




X<-c()
scores<-c()
for (i in 1:nrow(train_df_T)){

    if( train_df_T$Pclass[i]==1){
            score<-ifelse(train_df_T$SexCode[i]==1,0.5,0.01) * 1.7 *100
            #cat("\n1등급 좌석 점수:",score)
            scores<-append(scores,score) # 1등급 좌석일때 
    }
    else if(train_df_T$Pclass[i]==2){
            score<-(ifelse(train_df_T$SexCode[i]==1,0.5,0.01)) * 1.5 *100
            #cat("\n2등급 좌석 점수:",score)
            scores<-append(scores,score)# 2등급 좌석일때 
    }else{
        
        if(train_df_T$SexCode[i]==0 & train_df_T$Fare[i]<10  ){ #3등급 좌석이지만 남자일때 
            score<-(ifelse(train_df_T$SexCode[i]==1,0.5,0.01))*40
            #cat("\n3등급 남자 점수",score,"fare :",train_df_T$Fare[i])
            scores<-append(scores,score)#*ifelse(train_df_T$Fare[i]<150, 1,80)

        }else{
            score<-(ifelse(train_df_T$SexCode[i]==1,0.5,0.01))
            #cat("\n3등급 여자 점수 :",score)
            scores<-append(scores,score)  # 3등급 좌석일때 
        }
    }

}
cat("score 길이 :",length(scores),"/",nrow(train_df_T))
train_df_T$SPscore<-scores

#head(train_df_T,100)
#SP score 가 50 미만 인데 산사람 140


par(bg ='white')
#train_df_T[special_case,]

#hist(train_df[special_case,]$Fare)
train_df_T1<- train_df_T[,c("Survived","SPscore")]

rfFunc(train_df_T1,1,130,50)



pairs.panels(train_df_T1)
barplot(table(train_df_T1$Survived,train_df_T1$SPscore), beside = T)
subset(train_df_T,SPscore==0.5)
```

```r
# RandomForest 생사여부 예측률과 오분류표 구하기 
# RandomForrest Prediction
rfFunc <- function(data,target_col,rf_n,iter_n){
    Sys.setlocale(category = "LC_ALL", locale = "ko_KR.UTF-8") # 한글 설정
    target <- data[,target_col] # target 칼럼
    acc_rf <-c()
    acc_nn <-c()
    cat("예측률 \n")

    for(i in 1:iter_n){
        samp <- createDataPartition(data[,target_col], p =0.7, list =F)
        # 학습용, 검증용 분류
        data.tr     <- data[samp,]
        data.test   <- data[-samp,]
        # 문제 및 정답 만들기 
        x <- subset(data.test, select = -target_col)
        y <- data.test[,target_col]
        
        # 모델 생성
        target_name     <-colnames(data)[target_col] 
        Training_target <-data.frame(Survived=data.tr[,target_col])

        #model.nnet <- nnet(Training_target$V1 ~ ., data =data.tr, size = 3, trace =F) 

        rf <- randomForest(Training_target$Survived ~ ., data =data.tr, ntree =rf_n)
        # 예측
        pred_rf <-predict(rf,x )
        #pred_nn <-predict(model.nnet,x )
        acc_rf<-append(acc_rf,mean(pred_rf == y))
        #acc_nn<-append(acc_nn,mean(pred_nn == y))
        cat(mean(pred_rf ==y)," ")
    }
    cat("\n\033[36m Random Forest 평균 예측률 : ",mean(acc_rf) ,"\033[0m",sep = "")
    #cat("\n\033[36m Neural Network 평균 예측률 : ",mean(acc_nn) ,"\033[0m",sep = "")
    return(mean(acc_rf))
}
```
